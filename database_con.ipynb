```python
import csv
import sqlite3
```

```python
con = sqlite3.connect('Student_lifestyle.db')
cursor = con.cursor()
```

```python
create_table = """CREATE TABLE Students(
                    Student_ID INT PRIMARY KEY,
                    Study_Hours_Per_Day FLOAT,
                    Extracurricular_Hours_Per_Day FLOAT,
                    Sleep_Hours_Per_Day FLOAT,
                    Social_Hours_Per_Day FLOAT,
                    Physical_Activity_Hours_Per_Day FLOAT,
                    GPA FLOAT,
                    Stress_Level VARCHAR(20));
                """

cursor.execute(create_table)

```

```python
with open("student_lifestyle_dataset.csv") as f:
    data = csv.reader(f)
    header = next(data)  # Skip the header row
    for row in data:
        cursor.execute("""
        INSERT INTO Students VALUES (?, ?, ?, ?, ?, ?, ?, ?)
        """, (int(row[0]), float(row[1]), float(row[2]), float(row[3]), 
              float(row[4]), float(row[5]), float(row[6]), row[7]))
```
